```python
df['Loan_to_Income_Ratio'] = df['Loan_Amount'] / df['Income']
df
```

```python
df['Loan_to_Income_Ratio'] = df['Loan_to_Income_Ratio'].round(2)
df
```

```python
import numpy as np

df['Credit_Category'] = np.where(
    (df['Loan_to_Income_Ratio'] <= 4) & (df['Payment_History'] == 'Good'),
    'Good',
    np.where(
        ((df['Loan_to_Income_Ratio'] > 4) & (df['Loan_to_Income_Ratio'] <= 6)) | (df['Payment_History'] == 'Average'),
        'Average',
        'Poor'
    )
)

df
```

```python
df['Credit_Category'].value_counts()
```

```python
df.groupby('Credit_Category')['Loan_to_Income_Ratio'].mean()
```

```python
import matplotlib.pyplot as plt

df['Credit_Category'].value_counts().plot(kind='bar')
plt.title('Credit Category Distribution')
plt.xlabel('Credit Category')
plt.ylabel('Number of Customers')
plt.show()
```

```python
import numpy as np

df['Credit_Score'] = np.where(
    df['Credit_Category'] == 'Good',
    900 - (df['Loan_to_Income_Ratio'] * 10),
    np.where(
        df['Credit_Category'] == 'Average',
        700 - (df['Loan_to_Income_Ratio'] * 10),
        500 - (df['Loan_to_Income_Ratio'] * 10)
    )
)
df['Credit_Score'] = df['Credit_Score'].round(0)
df
```

```python
df['Risk_Segment'] = np.where(
    df['Credit_Score'] >= 750, 'Low Risk',
    np.where(
        df['Credit_Score'] >= 600, 'Medium Risk',
        'High Risk'
    )
)

df
```

```python
df['Risk_Segment'].value_counts()
```

```python
df.groupby('Risk_Segment')['Income'].mean()
```

```python
df.groupby('Risk_Segment')['Loan_Amount'].mean()
```

```python
import matplotlib.pyplot as plt

df['Risk_Segment'].value_counts().plot(kind='bar')
plt.title('Customer Risk Distribution')
plt.xlabel('Risk Segment')
plt.ylabel('Customer Count')
plt.show()
```
